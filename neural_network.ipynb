```python
from tensorflow import keras
# 2. Baseline ANN model
def build_baseline_model():
    model = keras.Sequential()
    model.add(keras.layers.Dense(64, input_dim=16, activation='relu'))
    model.add(keras.layers.Dense(32, activation='relu'))
    model.add(keras.layers.Dense(26, activation='softmax'))
    model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
baseline_model = build_baseline_model()
history = baseline_model.fit(x_train,y_train, epochs=100, batch_size = 32, validation_split=0.2, verbose=0  )
```

```python
# Evaluate baseline
baseline_pred = baseline_model.predict(x_test)
baseline_pred = np.argmax(baseline_pred, axis=1)
from sklearn.metrics import accuracy_score
baseline_acc = accuracy_score(y_test, baseline_pred)
print("Baseline Model Accuracy: {:.2f}%".format(baseline_acc * 10))
```

3. Hyperparameter Tuning

Modify various hyperparameters, such as the number of hidden layers, neurons per hidden layer, activation functions, and learning rate, to observe their impact on model performance.

```python
!pip install --upgrade --quiet keras_tuner
```

```python
def build_model(hp):
    model = keras.Sequential()
    model.add(layers.InputLayer(input_shape=(x_train.shape[1],)))
    # Tune number of layers
    for i in range(hp.Int('num_layers', 1, 3)):
        model.add(layers.Dense(
            units=hp.Int(f'units_{i}', min_value=32, max_value=256, step=32),
            activation=hp.Choice('activation', ['relu', 'tanh'])
        ))
    model.add(layers.Dense(len(np.unique(y)), activation='softmax'))
    # Tune learning rate
    lr = hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=lr),
                  loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

Adopt a structured approach like grid search or random search for hyperparameter tuning, documenting your methodology thoroughly.

```python
# Random search tuner
import keras_tuner as kt
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=1,
    directory='tuner_logs',
    project_name='alphabets_ann'
)

tuner.search(x_train, y_train, epochs=10, validation_split=0.2, verbose=0)

best_hps = tuner.get_best_hyperparameters(1)[0]
```

4. Evaluation

Employ suitable metrics such as accuracy, precision, recall, and F1-score to evaluate your model's performance.

Discuss the performance differences between the model with default hyperparameters and the tuned model, emphasizing the effects of hyperparameter tuning.

```python
# Build tuned model
tuned_model = tuner.hypermodel.build(best_hps)
tuned_model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2, verbose=0)

# Evaluate tuned model
tuned_preds = np.argmax(tuned_model.predict(x_test), axis=-1)
print("\nTuned Model Performance:")
print("Accuracy:", accuracy_score(y_test, tuned_preds))
print(classification_report(y_test, tuned_preds, target_names=le.classes_))
```
